```python
from scipy.signal import correlate
from scipy.stats import zscore

%matplotlib inline
rs = int((250 - GRB221009A_prompr.t0) / GRB221009A_prompr.dt)
re = int((275 - GRB221009A_prompr.t0) / GRB221009A_prompr.dt)

# Standardize the data
data = zscore(GRB221009A_prompr.counts[rs:re])

# Define the step function
def step_function(x, position, width):
    return np.where((x >= position) & (x <= position + width), 1, 0)

# Define the range for position and width with a step of 0.01
positions = np.arange(250, 276, 0.1)
widths = np.arange(0, 21, 0.1)

# Initialize the correlation matrix
correlation_matrix = np.zeros((len(positions), len(widths)))

# Scan through the positions and widths
for i, pos in enumerate(positions):
    for j, width in enumerate(widths):
        # Generate the step function
        step = step_function(GRB221009A_prompr.time[rs:re], pos, width)
        # Standardize the step function
        step = zscore(step)
        # Compute the correlation
        correlation = correlate(data, step, mode='valid')
        # Store the maximum correlation value
        correlation_matrix[i, j] = np.max(correlation)

# Find the index of the maximum correlation value in the matrix
max_corr_idx = np.unravel_index(np.nanargmax(correlation_matrix), correlation_matrix.shape)
best_position = positions[max_corr_idx[0]]
best_width = widths[max_corr_idx[1]]
print(best_position, best_width)

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, aspect='auto', extent=[0, 20, 275, 250], origin='upper', cmap='viridis')
plt.colorbar(label='Correlation Coefficient')
plt.xlabel('Width')
plt.ylabel('Position')
plt.title('Standardized Cross-Correlation')
plt.scatter([best_width], [best_position], color='red', marker='x', label=f'Best Position: {best_position}, Width: {best_width}')
plt.legend()
plt.show()

plt.figure(figsize=(10, 8))
plt.plot(GRB221009A_prompr.time[rs:re], GRB221009A_prompr.counts[rs:re])
best_step = 10000 * step_function(GRB221009A_prompr.time[rs:re], best_position, best_width)
plt.plot(GRB221009A_prompr.time[rs:re], best_step, '--', label=f'Best Step Function: pos={best_position}, width={best_width}')
plt.legend()
plt.show()
```

```python
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt

# 提供更好的初始参数
initial_params = [10000, 261, 3, 10]

# 定义拟合函数
def fit_function(params, x, y):
    error = np.sum((y - my.step(x, *params))**2)
    print(f"Params: {params}, Error: {error}")
    return error

rs = int((245 - GRB221009A_prompr.t0) / GRB221009A_prompr.dt)
re = int((300 - GRB221009A_prompr.t0) / GRB221009A_prompr.dt)

# 使用minimize进行拟合，使用L-BFGS-B算法
result = minimize(fit_function, initial_params, args=(GRB221009A_prompr.time[rs:re], GRB221009A_prompr.counts[rs:re]), method='L-BFGS-B', options={'maxiter': 10000000, 'ftol': 1e-20})

params = result.x

plt.plot(GRB221009A_prompr.time[rs:re], GRB221009A_prompr.counts[rs:re])
xx = np.arange(245, 300, 0.1)
ll = ""
for i, par in enumerate(params):
    ll += f"par{i+1}: {par:.2f} . "
plt.plot(xx, my.step(xx, *params), label=ll)
plt.legend()
plt.show()
```
