```python
import os
os.environ['OPENBLAS_NUM_THREADS'] = '1'

import ROOT

import numpy as np
import matplotlib.pyplot as plt

from astropy.io import fits
from astropy.wcs import WCS
from astropy.coordinates import SkyCoord
from astropy import units as u
from astropy.visualization import astropy_mpl_style, imshow_norm
from astropy.coordinates import Angle

import mylib as my
import root_numpy as rt
from importlib import reload
```

```python
%matplotlib inline
name="J0057"

root_file=ROOT.TFile.Open(("../../data/gll_dust.root"),"read")
root_th2d=root_file.Get("gll_region")
X_nbins=root_th2d.GetNbinsX()
Y_nbins=root_th2d.GetNbinsY()
X_min=root_th2d.GetXaxis().GetXmin()
X_max=root_th2d.GetXaxis().GetXmax()
Y_min=root_th2d.GetYaxis().GetXmin()
Y_max=root_th2d.GetYaxis().GetXmax()
X_size=(X_max-X_min)/X_nbins
Y_size=(Y_max-Y_min)/Y_nbins
print(X_min,X_max,X_nbins, X_size)
print(Y_min,Y_max,Y_nbins, Y_size)
data = rt.hist2array(root_th2d).T
plt.imshow(data,aspect="auto",extent=[5,235,-20,20],origin='lower')
plt.colorbar()
```

```python
ra_G25, dec_G25 =  14, 63.5
lranges = 5
branges = 5
ra1,dec1=ra_G25, dec_G25
l,b = my.edm2gal(ra1,dec1)
l=int(l); b=int(b)
ll = np.arange(l-lranges,l+lranges,X_size)
bb =  np.arange(-branges,branges,Y_size)
L,B = np.meshgrid(ll,bb)
RA, DEC = my.gal2edm(L,B)
```

```python
%matplotlib inline
lrange=[l-lranges,l+lranges]
brange=[-branges,branges]
dataneed = data[int((brange[0]-Y_min)/Y_size):int((brange[1]-Y_min)/Y_size),int((lrange[0]-X_min)/X_size):int((lrange[1]-X_min)/X_size)]
plt.imshow(np.log(dataneed),aspect="auto",extent=[lrange[0],lrange[1],brange[0],brange[1]],origin='lower')
plt.colorbar()

plt.xlabel("Galactic Longitude (deg)")
plt.ylabel("Galactic Latitude (deg)")
plt.scatter(l,b,marker='x', color='red')
plt.annotate(name, xy=(l,b), xytext=(l+0.5,b+0.5),color='red')
# plt.gca().invert_xaxis()

plt.figure()
plt.scatter(RA,DEC,c=np.log(dataneed),s=5)
plt.scatter(ra1,dec1,marker='x', color='red')
plt.annotate("J0057", xy=(ra1,dec1), xytext=(ra1+0.5,dec1+0.5),color='red')
# plt.ylim(58,68)
# plt.xlim(0,50)
```

```python
s = dataneed.copy()
for idec,decd in  enumerate(dataneed):
    ddd = brange[0]+idec*Y_size
    for ira,counts in enumerate(decd):
        lll = lrange[0]+ira*X_size
        s[idec,ira] = (np.radians(X_size)*np.radians(Y_size)*np.cos(np.radians(ddd)))
plt.imshow(np.log(s),aspect="auto",extent=[lrange[0],lrange[1],brange[0],brange[1]],origin='lower')
plt.colorbar()
A = np.multiply(dataneed,s)
ss = np.sum(s)
sa = np.sum(A)
print(ss)
print(sa)
plt.figure()
plt.imshow(np.log(A),aspect="auto",extent=[lrange[0],lrange[1],brange[0],brange[1]],origin='lower')
plt.colorbar()
```

```python
# 
naxis1 = len(ll)  # 
naxis2 = len(bb)  # 

# 
lon_range = lrange  # 
lat_range = brange  # 

#  WCS 
wcs = WCS(naxis=2)
wcs.wcs.crpix = [naxis1 / 2, naxis2 / 2]  # 
wcs.wcs.cdelt = np.array([0.1, 0.1])  # 
wcs.wcs.crval = [l, b]  # 
wcs.wcs.ctype = ['GLON-CAR', 'GLAT-CAR']  # 
# 
header = fits.Header()
header.update(wcs.to_header())
header['OBJECT'] = 'Test Image'
header['BUNIT'] = 'Jy/beam'

#  HDU
hdu = fits.PrimaryHDU(data=dataneed/sa, header=header)

#  FITS 
hdu.writeto(f'./{name}_dust_bkg_template.fits', overwrite=True)
```
